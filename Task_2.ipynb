```python
train_labels = train_df.iloc[:, 1:].values
test_labels = test_df.iloc[:, 1:].values
```

```python
model=keras.Sequential()
model.add(keras.Input(shape=(9,)))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
print(model.summary())
```

```python
model.compile(loss=keras.losses.MeanAbsoluteError(),
             optimizer=keras.optimizers.Adam(learning_rate=0.001),
             metrics=['accuracy'])
```

```python
model.fit(train_df,train_labels,batch_size=32,epochs=5,verbose=2)
```

```python
model.evaluate(test_df,test_labels,batch_size=32,verbose=2)
```
